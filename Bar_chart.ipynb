```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGLE','AMZN','MFST','FB']
revenue=[90,136,89,120]
profit=[40,20,34,50]
```

```python
# For convert string_name as array index to show in graph
ypos = np.arange(len(company)) 
ypos
```

```python
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.bar(ypos,revenue)
```

```python

plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("BD Tech Stocks")

plt.bar(ypos,revenue,label="Revenue")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("BD Tech Stocks")

plt.bar(ypos,revenue,label="Revenue")
plt.bar(ypos,profit,label="profit")

plt.legend(loc="best")
```

```python
xpos = np.arange(len(company)) 
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("BD Tech Stocks")

plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")

plt.legend(loc="best")
```
