# one hotencoding using Loop

```python
df7.head()
```

```python
cols = ['Area']
for col in cols:
    dummy = pd.get_dummies(df7[col],drop_first = True,prefix='Area')
    df8 = pd.concat([df7,dummy],axis=1).drop(col,axis=1)
df8.head()    
```
