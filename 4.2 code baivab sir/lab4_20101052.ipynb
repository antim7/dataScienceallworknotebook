```python
class SymbolTable:
    def __init__(self):
        self.table = []

    def insert(self, label, type_, size, dimension, line_of_code, address):
        self.table.append((label, type_, size, dimension, line_of_code, address))

    def search(self, label, type_):
        for entry in self.table:
            if entry[0] == label and entry[1] == type_:
                return entry
        return None

    def delete(self, label, type_):
        for entry in self.table:
            if entry[0] == label and entry[1] == type_:
                self.table.remove(entry)
                return True
        return False

    def update(self, label, type_, new_entry):
        for i, entry in enumerate(self.table):
            if entry[0] == label and entry[1] == type_:
                self.table[i] = new_entry
                return True
        return False

    def display(self):
        print("Label | Type | Size | Dimension | Line of Code | Address")
        print("-" * 55)
        for entry in self.table:
            print(" | ".join(entry))

def main():
    symbol_table = SymbolTable()

    while True:
        print("\nOptions:")
        print("1. Insert")
        print("2. Search")
        print("3. Delete")
        print("4. Update")
        print("5. Display")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            label = input("Enter Label: ")
            type_ = input("Enter Type: ")
            size = input("Enter Size: ")
            dimension = input("Enter Dimension: ")
            line_of_code = input("Enter Line of Code: ")
            address = input("Enter Address: ")
            symbol_table.insert(label, type_, size, dimension, line_of_code, address)

        elif choice == '2':
            label = input("Enter Label: ")
            type_ = input("Enter Type: ")
            result = symbol_table.search(label, type_)
            if result:
                print("Symbol found:", result)
            else:
                print("Symbol not found.")

        elif choice == '3':
            label = input("Enter Label: ")
            type_ = input("Enter Type: ")
            if symbol_table.delete(label, type_):
                print("Symbol deleted successfully.")
            else:
                print("Symbol not found.")

        elif choice == '4':
            label = input("Enter Label: ")
            type_ = input("Enter Type: ")
            new_entry = (label, type_, input("Enter Size: "), input("Enter Dimension: "), input("Enter Line of Code: "), input("Enter Address: "))
            if symbol_table.update(label, type_, new_entry):
                print("Symbol updated successfully.")
            else:
                print("Symbol not found.")

        elif choice == '5':
            symbol_table.display()

        elif choice == '6':
            print("Exiting...")
            break

        else:
            print("Invalid choice. Please choose again.")

if __name__ == "__main__":
    main()
```
